## Visualization using Python Matplotlib Library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 1. Plot basic plot
Days = [1,2,3,4,5,6,7,8]
Speed = [60,62,61,58,56,57,46,63]

plt.plot(Days, Speed)
plt.title("Car Speed Measurement")
plt.xlabel("Days")
plt.ylabel("Car Speed")
plt.show()
```

```python
# 2. Apply some string formats like line style ex green dotted line, marker shape like +,
# change markersze, markerface color, etc
# ms = marker size, mfc = marker face color, mec = marker border color
plt.plot(Days, Speed, marker='+',ms=25,mec='red',mfc='green', linestyle='dotted', linewidth=3)
plt.title("Car Speed Measurement")
plt.xlabel("Days")
plt.ylabel("Car Speed")
plt.show()
```

```python
# 3. Plot Axes Labels, Chart title, Legend, Grid in Car minimum, Maximum and average speed in 8 days.

days=[1,2,3,4,5,6,7,8]
max_speed=[80,91,92,88,77,79,76,75]
min_speed=[42,43,40,42,33,36,34,35]
avg_speed=[46,58,57,56,40,42,41,36]

plt.figure(figsize=(8,6))
plt.subplot(3,1,1)
plt.plot(days, max_speed, color = 'red', marker = '*', ms = 10, mec = 'red', mfc = 'orange', label = 'Maximum Speed')
plt.xlabel('Days')
plt.ylabel('Maximum Speed')
plt.title('Car Speed Measurement')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(8,6))
plt.subplot(3,1,2)
plt.plot(days, min_speed, color = 'green', marker = 'o', ms = 10, mec = 'black', mfc = 'blue', label = 'Minimum Speed')
plt.xlabel('Days')
plt.ylabel('Minimum Speed')
plt.title('Car Speed Measurement')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(8,6))
plt.subplot(3,1,3)
plt.plot(days, avg_speed, color = 'yellow', marker = '^', ms = 10, mec = 'red', mfc = 'orange', label = 'Average Speed')
plt.xlabel('Days')
plt.ylabel('Average Speed')
plt.title('Car Speed Measurement')
plt.legend()
plt.grid()
plt.show()
```

```python
# 4. Plotting a basic sine graph by adding more features. 
# Adding Multiple plots by Superimposition like cosine wave.
x = np.arange(0,4*np.pi,0.1)  
y = np.sin(x)
z = np.cos(x)

plt.plot(x,y, color = 'green', marker='o')
plt.plot(x,z, color = 'blue', marker = '*')
plt.xlabel('Sin and Cos Value From 0 to 4pi')
plt.ylabel('Sin(x) and Cos(x)')
plt.title('Sin And Cosine Wave')
plt.legend(['Sin(x)','Cos(x)'])
plt.show()
```

```python
# 5. Plot Simple bar chart showing popularity of Programming Languages.
# Plot Multiple Bars showing Popularity and Security of major Programming Languages. 
# Also Create Horizontal bar chart using barh function.
Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']
Popularity = [56, 39, 34, 34, 29]
Security = [44 ,36 ,55, 50, 42]

n = len(Languages)
r = np.arange(n)
width = 0.25

plt.bar(r, Popularity, width=width, color = 'blue', label = 'Popularity')
plt.bar(r+width, Security, width=width, color = 'red', label = 'Security')

plt.xlabel('Languages')
plt.xticks(r+width/2,Languages)
plt.ylabel('Score')
plt.title('Popularity of Programming Languages')
plt.legend()
plt.show()
```

```python
# 6. Plot Histogram, We have a sample data of Students marks of various Students, we will try to plot number of Students 
# by marks range and try to figure out how many Students are average, below-average and Excellent.

marks = np.array([61,86,42,46,73,95,65,78,53,92,55,69,70,49,72,86,64])

below_avg = marks[np.logical_and(marks >= 40, marks < 60)]
avg_marks = marks[np.logical_and(marks >= 60, marks < 80)]
excel_marks = marks[np.logical_and(marks >= 80, marks <= 100)]

plt.hist(marks)
plt.show()
```

```python
print('Total Below Average Students Are : ', below_avg.size)
print('Total Average Students Are : ', avg_marks.size)
print('Total Excellent Students Are : ', excel_marks.size)
```

```python
# 7.  Titanic Data Set
# (i) Create a pie chart presenting the male/female proportion 
# (ii) Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

titanic = pd.read_csv('./datasets/titanic.csv')
titanic.head()
```

```python
# (i) Create a pie chart presenting the male/female proportion 
titanic['Sex']

mf = titanic.Sex.value_counts()

mf

plt.pie(mf, autopct='%.0f%%')

plt.legend(['Male','Female'])
plt.title('Male And Female Proportion')
plt.show()
```

```python
# (ii) Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

male = titanic[titanic['Sex']=='male']
male.Sex.value_counts()

female = titanic[titanic['Sex']=='female']
female.Sex.value_counts()

plt.scatter(male.Fare, male.Age, label = 'Male', marker = '*', color='blue')

plt.scatter(female.Fare, female.Age, label = 'Female', marker = 'o', color='red')

plt.legend()
plt.xlabel('Fare')
plt.ylabel('Age')
plt.grid()
plt.show()
```
